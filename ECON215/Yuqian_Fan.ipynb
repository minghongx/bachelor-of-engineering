# COVID-19's Impact on UK Airport Traffic

## Part 1
### Some impotant airport datas in 2018-2021 

````python
import numpy as np
import pandas as pd
np.random.seed(12345)
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(15, 10))
from pathlib import Path
from datetime import datetime


tables_path1 = Path().rglob("*-*/*Aircraft_Movements*") 
#create a list of paths containing all the monthly 'Aircraft_Movements.csv' from 2018-2021downloaded beforehand
tables_path2 = Path().rglob("*-*/*Terminal_and_Transit_Passengers*")#the same
tables_path3 = Path().rglob("*-*/*Freight_by_Type_and_Nationality*")
tables_path4 = Path().rglob("*-*/*Mail_by_Type_and_Nationality*")

frame1 = [pd.read_csv(table_path).groupby(["reporting_period"])[["grand_total"]].sum() for table_path in tables_path1]
#read all the monthly 'Aircraft_Movements.csv' from 2018-2021 and sum the aircraft movements in each month
frame2 = [pd.read_csv(table_path).groupby(["this_period"])[["total_pax_this_period"]].sum() for table_path in tables_path2]
frame3 = [pd.read_csv(table_path).groupby(["reporting_period"])[["total_freight"]].sum() for table_path in tables_path3]
frame4 = [pd.read_csv(table_path).groupby(["reporting_period"])[["total_mail"]].sum() for table_path in tables_path4]

results = [pd.concat(frame) for frame in [frame1,frame2,frame3,frame4]]
#creat a list containing 4 dataframes, concatenated from frame1-4 

for result in results:
    result.index = pd.to_datetime(result.index, format="%Y%m")
    # convert the index of each dataframe in results to datetime
    result.index.name = 'dates'
    # rename the index name, making it more straightforward


DataFrame1=pd.concat(results, axis=1)
#combine the 4 dataframe in results on axis=1
DataFrame1.columns = ['Aircraft movements', 'Passengers', 'Freights', 'Mails']
#rename the column names
print(DataFrame1)
````

````python
fig1= plt.figure()

ax1 = fig1.add_subplot(2, 2, 1) #plot on ax1 the changes of aircraft movements in 2018-2021
ax1.plot(results[0].index, results[0].values, 'k^--')
ax1.set_ylabel('Number of aircraft movements')
ax1.set_xlabel('date')
ax1.set_title('2018-2021 Aircraft_Movements')
plt.xticks(rotation=45)

ax2 = fig1.add_subplot(2, 2, 2)#the same
ax2.plot(results[1].index, results[1].values, 'k^--')
ax2.set_ylabel('Number of passengers')
ax2.set_xlabel('date')
ax2.set_title('2018-2021 Terminal_and_Transit_Passengers')
plt.xticks(rotation=45)

ax3 = fig1.add_subplot(2, 2, 3)
ax3.plot(results[2].index, results[2].values, 'k^--')
ax3.set_ylabel('Freights/Tonne')
ax3.set_xlabel('date')
ax3.set_title('2018-2021 Airtransportation Freights')
plt.xticks(rotation=45)

ax4 = fig1.add_subplot(2, 2, 4)
ax4.plot(results[3].index, results[3].values, 'k^--')
ax4.set_ylabel('Mails/Tonne')
ax4.set_xlabel('date')
ax4.set_title('2018-2021 Airtransportation Mails')
plt.xticks(rotation=45)

COVID_data=[
    (datetime(2020,1,29), 'first case diagnosed in the UK'),
    (datetime(2020,12,8), 'first COVID vaccine jab in UK'),
    (datetime(2020,10,1), 'mutated virus Dleta found in India')
]
#I found some important COVID data online
ax1.annotate(COVID_data[0][1], 
            xy=(COVID_data[0][0], 2.1e5),
            xytext=(COVID_data[0][0], 3.1e5),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='center', 
            verticalalignment='top',
            fontsize= 9)
#add COVID_data to ax1

ax1.annotate(COVID_data[1][1], 
            xy=(COVID_data[1][0], 1e5),
            xytext=(COVID_data[1][0], 2e5),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='center', 
            verticalalignment='top',
            fontsize= 9)
#the same
ax1.annotate(COVID_data[2][1], 
            xy=(COVID_data[2][0], 1.5e5),
            xytext=(COVID_data[2][0], 2.5e5),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='center', 
            verticalalignment='top',
            fontsize= 9)




ax2.annotate(COVID_data[0][1], 
            xy=(COVID_data[0][0], 2e7),
            xytext=(COVID_data[0][0], 3e7),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='center', 
            verticalalignment='top',
            fontsize= 9)

ax2.annotate(COVID_data[1][1], 
            xy=(COVID_data[1][0], 0.3e7),
            xytext=(COVID_data[1][0], 1.3e7),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='center', 
            verticalalignment='top',
            fontsize= 9)

ax2.annotate(COVID_data[2][1], 
            xy=(COVID_data[2][0], 0.5e7),
            xytext=(COVID_data[2][0], 1.5e7),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='center', 
            verticalalignment='top',
            fontsize= 9)



ax3.annotate(COVID_data[0][1], 
            xy=(COVID_data[0][0], 1.9e5),
            xytext=(COVID_data[0][0], 2.4e5),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='center', 
            verticalalignment='top',
            fontsize= 9)

ax3.annotate(COVID_data[1][1], 
            xy=(COVID_data[1][0], 2e5),
            xytext=(COVID_data[1][0], 2.2e5),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='center', 
            verticalalignment='top',
            fontsize= 9)

ax3.annotate(COVID_data[2][1], 
            xy=(COVID_data[2][0], 1.9e5),
            xytext=(COVID_data[2][0], 2.3e5),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='center', 
            verticalalignment='top',
            fontsize= 9)  


ax4.annotate(COVID_data[0][1], 
            xy=(COVID_data[0][0], 1.5e4),
            xytext=(COVID_data[0][0], 1.8e4),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='center', 
            verticalalignment='top',
            fontsize= 9)

ax4.annotate(COVID_data[1][1], 
            xy=(COVID_data[1][0], 1.7e4),
            xytext=(COVID_data[1][0], 1.85e4),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='center', 
            verticalalignment='top',
            fontsize= 9)

ax4.annotate(COVID_data[2][1], 
            xy=(COVID_data[2][0], 1.49e4),
            xytext=(COVID_data[2][0], 1.7e4),
            arrowprops=dict(facecolor='black', headwidth=4, width=2,
                            headlength=4),
            horizontalalignment='center', 
            verticalalignment='top',
            fontsize= 9)          


fig1.subplots_adjust(hspace=0.80, wspace=0.60) #adjust the space btween graphs


plt.show()
````
